```python
from docling.document_converter import DocumentConverter
from markitdown import MarkItDown
import pdfplumber
```

```python
pdf_path = "../data/raw/ch591_noise.pdf"
```

### PDF Plumber

```python
pdfplumber_text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages):
        raw_text = page.extract_text()
        pdfplumber_text += f"Page {page_num}:\n{raw_text}\n\n"

print("PDFPlumber Extraction:")
print(pdfplumber_text)
```

### MarkItDown

```python
markitdown_text = MarkItDown().convert(pdf_path).text_content
print("MarkItDown Extraction:")
print(markitdown_text)
```

### Docling

```python
converter = DocumentConverter()
result = converter.convert(pdf_path)
docling_text = result.document.export_to_markdown()

print("Docling Extraction:")
print(docling_text)
```

### Thoughts on Comparison of Parsers

- MarkItDown cannot extract table formatting or list formatting. It is ideal for when extracting raw text irrespective of the formatting. This does not work well for Policy Documents wherein we want to group information by sections for citation purposes.

- Docling is good in that it highlights headings and subheadings with a # symbol. However, it also extracts headers and footers for every page. It is a bit too noisy. 

- PdfPlumber is the best option for this usecase.
